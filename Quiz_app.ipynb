```python
import tkinter as tk
from tkinter import messagebox

# Declare login entry fields as global variables.
entry_username = None
entry_password = None

# Function to handle login action.
def handle_login():
    # Access global variables for username and password.
    global entry_username, entry_password
    username = entry_username.get()
    password = entry_password.get()

    # Simple check (you can replace this with a database or more complex logic).
    if username == "Python1991" and password == "February20":
        messagebox.showinfo("Login Successful", "Welcome to the Quiz Application!")
        open_quiz_window()  # Proceed to quiz window.
        window.quit()  # Close login window.
    else:
        messagebox.showerror("Login Failed", "Invalid username or password")

# Function to open quiz window.
def open_quiz_window():
    # Create new window for quiz.
    quiz_window = tk.Tk()
    quiz_window.title("Quiz")
    quiz_window.geometry("1600x1000")

    # Set background color of quiz window.
    quiz_window.config(bg="lightblue")

    # Some question as a sample questions and answers.
    questions = [
        {"question": "Which of the following is the correct syntax to create a function in Python?", "options": ["create function():", "function def():", "def function():", "function create():"], "answer": "def function():"},
        {"question": "Which method can be used to add an element to the end of a list in Python?", "options": ["add()", "append()", "insert()", "extend()"], "answer": "append()"},
        {"question": "Which of the following is used to define a block of code in Python?", "options": ["Parentheses()", "Indentation (spaces or tabs)", "Curly braces {}", "Square brackets []"], "answer": "Indentation (spaces or tabs)"},
        {"question": "What does the len() function do?", "options": ["Returns the number of characters in a string", "Returns the length of a list, tuple, or string", "Returns the sum of the elements in a list", "Returns the last element of a list"], "answer": "Returns the length of a list, tuple, or string"},
        {"question": "Which of the following is a valid way to comment a single line in Python?", "options": ["/* This is a comment */", "# This is a comment", "// This is a comment", "<!-- This is a comment -->"], "answer": "# This is a comment"},
        {"question": "Which operator is used to compare two values in Python?", "options": ["=", "==", ":=", "!="], "answer": "=="},
        {"question": "Which of the following methods can be used to remove an item from a list in Python?", "options": ["remove()", "delete()", "pop()", "Both a and c"], "answer": "Both a and c"},
    ]
    
    score = 0  # Initialize score.
    current_question_index = 0  # Track the current question.

    # Function check selected answer.
    def check_answer(selected_answer, correct_answer):
        nonlocal score
        if selected_answer == correct_answer:
            score += 1
        next_question()

    # Function show next question.
    def next_question():
        nonlocal current_question_index
        if current_question_index < len(questions):
            question_data = questions[current_question_index]
            question_label.config(text=question_data["question"])

            # Set random colors each option.
            option_colors = ["lightcoral", "lightgreen", "lightyellow", "lightblue"]  # Different colors for each option.

            for i, option in enumerate(question_data["options"]):
                option_buttons[i].config(
                    text=option,
                    command=lambda option=option: check_answer(option, question_data["answer"]),
                    bg=option_colors[i]  # Apply the color.
                )

            current_question_index += 1
        else:
            # Show final score and close the quiz window.
            messagebox.showinfo("Quiz Finished", f"Your score is: {score}/{len(questions)}")
            quiz_window.quit()  # Close the quiz window.

    # Create label for question with wraplength.
    question_label = tk.Label(quiz_window, text="", font=("Arial", 20), width=80, wraplength=700, bg="lightblue")
    question_label.pack(pady=20)

    # Create buttons for answer options with larger size.
    option_buttons = []
    for _ in range(4):
        button = tk.Button(quiz_window, font=("Arial", 12), width=30, height=2)  # Increased width and height
        button.pack(pady=10)
        option_buttons.append(button)

    # Start quiz with first question.
    next_question()

    # Start tkinter event loop for quiz window.
    quiz_window.mainloop()

# Function open login window.
def open_login_window():
    global window, entry_username, entry_password
    window = tk.Tk()
    window.title("Quiz Application Login")
    window.geometry("1000x600")

    # Choose background color for login window.
    window.config(bg="lightgreen")

    # Add labels and entry fields for username and password.
    label_username = tk.Label(window, text="Username:", bg="lightgreen")
    label_username.pack(pady=10)

    entry_username = tk.Entry(window)
    entry_username.pack(pady=5)

    label_password = tk.Label(window, text="Password:", bg="lightgreen")
    label_password.pack(pady=10)

    entry_password = tk.Entry(window, show="*")  # 'show' hide the password input.
    entry_password.pack(pady=5)

    # Add login button with green background.
    login_button = tk.Button(window, text="Login", command=handle_login, bg="green", fg="white")
    login_button.pack(pady=10)

    # Start Tkinter event loop for login window.
    window.mainloop()

# Start a application by opening login window.
open_login_window()
```
